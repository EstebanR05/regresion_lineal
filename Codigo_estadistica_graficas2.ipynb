```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# Datos de la tabla
X = np.array([2926, 3259, 2660, 3094, 2930, 2895, 2844, 3438, 1921, 2266, 3038, 2130, 3282, 1887, 3196, 2923, 2671, 3487, 1930, 3485]).reshape(-1, 1)
Y = np.array([70.72, 72.33, 68.69, 70.37, 68.91, 69.91, 70.57, 71.76, 63.87, 65.76, 74.67, 64.60, 71.40, 68.95, 66.29, 73.82, 71.03, 71.11, 66.90, 75.51])

# Ajuste del modelo de regresión lineal
model = LinearRegression()
model.fit(X, Y)

# Coeficientes del modelo
a = model.intercept_  # Intercepto
b = model.coef_[0]    # Pendiente

# Predicciones
Y_pred = model.predict(X)

# Métricas de evaluación
r_squared = model.score(X, Y)  # Coeficiente de determinación R^2
rmse = np.sqrt(mean_squared_error(Y, Y_pred))
mae = mean_absolute_error(Y, Y_pred)

# Resultados
print("Ecuación del modelo: Y = {:.4f} + {:.4f}X".format(a, b))
print("Coeficiente de determinación (R^2): {:.4f}".format(r_squared))
print("Error cuadrático medio (RMSE): {:.4f}".format(rmse))
print("Error absoluto medio (MAE): {:.4f}".format(mae))

# Gráfica del ajuste
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Datos reales')
plt.plot(X, Y_pred, color='red', label='Modelo ajustado')
plt.title('Regresión Lineal: Consumo Calórico vs Peso Corporal')
plt.xlabel('Consumo Calórico (X)')
plt.ylabel('Peso Corporal (Y)')
plt.legend()
plt.show()

# Análisis de residuos
residuos = Y - Y_pred
plt.figure(figsize=(10, 6))
plt.scatter(Y_pred, residuos, color='purple')
plt.axhline(0, color='black', linestyle='--')
plt.title('Análisis de Residuos')
plt.xlabel('Valores Predichos')
plt.ylabel('Residuos')
plt.show()
```
