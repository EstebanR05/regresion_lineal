```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Datos predefinidos
def cargar_datos():
    print("Cargando datos predefinidos...")
    data = pd.DataFrame({
        'X1': [2926, 3259, 2660, 3094, 2930, 2895, 2844, 3438, 1921, 2266,
               3038, 2130, 3282, 1887, 3196, 2923, 2671, 3487, 1930, 3485],
        'Y': [70.72, 72.33, 68.69, 70.37, 68.91, 69.91, 70.57, 71.76, 63.87,
              65.76, 74.67, 64.60, 71.40, 68.95, 66.29, 73.82, 71.03, 71.11,
              66.90, 75.51]
    })
    print("\nDatos cargados:")
    print(data)
    return data

# Cargar los datos
data = cargar_datos()

# Separar variables independientes y dependientes
X = data[['X1']]
Y = data['Y']

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, Y)

# Obtener coeficientes del modelo
intercepto = model.intercept_
pendiente = model.coef_[0]
print("\nEcuación del modelo: Y = {:.4f} + {:.4f}X".format(intercepto, pendiente))

# Predicciones y evaluación del modelo
Y_pred = model.predict(X)
r2 = model.score(X, Y)  # Usar el método `score` del modelo para R²
rmse = np.sqrt(mean_squared_error(Y, Y_pred))
mae = mean_absolute_error(Y, Y_pred)
print("\nMétricas del modelo:")
print(f"R^2: {r2:.4f}")  # Debería coincidir con el script original
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")

# Gráfica del ajuste del modelo
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Datos reales')
plt.plot(X, Y_pred, color='red', label='Modelo ajustado')
plt.title('Regresión Lineal: X1 vs Y')
plt.xlabel('X1 (Consumo Calórico)')
plt.ylabel('Y (Peso Corporal)')
plt.legend()
plt.show()

# Análisis de residuos
residuos = Y - Y_pred
plt.figure(figsize=(8, 5))
sns.histplot(residuos, kde=True, color='purple')
plt.title('Distribución de los Residuos')
plt.xlabel('Residuos')
plt.show()
```
